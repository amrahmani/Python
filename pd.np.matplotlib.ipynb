```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 8, 5])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 8, 5], marker='x') # Added a marker for visibility
plt.title("Plot starting at x=0 on the axis")
plt.xlabel("X-axis (Starts at 0)")
plt.ylabel("Y-axis (Starts at 0)")
plt.xlim(xmin=0) # Ensures the x-axis starts at 0
plt.ylim(ymin=0)
plt.grid(True)
plt.savefig("myplot.png")
```
