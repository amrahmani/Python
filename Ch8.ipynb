```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the x-axis (e.g., time in seconds)
x = [0, 1, 2, 3, 4, 5]

# Sample data for the y-axis (e.g., some measured value)
y = [10, 12, 8, 15, 11, 18]

# Create the line plot
plt.plot(x, y)

# Add labels to the x and y axes
plt.xlabel("Time (seconds)")
plt.ylabel("Measured Value")

# Add a title to the graph
plt.title("Trend of Measured Value Over Time")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the scatter plot
x_values = [1, 5, 2, 8, 3, 9]
# x_values = np.array([1, 5, 2, 8, 3, 9])

y_values = [3, 7, 4, 6, 1, 8]
# y_values = np.array([3, 7, 4, 6, 1, 8])

# Create the scatter plot
plt.scatter(x_values, y_values)

# Add labels and title (optional, but good practice)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example with custom colors and labels
categories = ["Sydney", "Adelaide", "Melbourne", "Perth"]
values = [3, 8, 6, 10]
colors = ['skyblue', 'red', 'lightgreen', 'gold']

plt.figure(figsize=(8, 6)) # Adjust figure size for
                           # better readability

plt.bar(categories, values, color=colors)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot with Custom Colors and Labels")

plt.show()
```
