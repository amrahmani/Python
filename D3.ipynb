```python
%%html
<div id="chart"></div>

<script src="https://cdn.jsdelivr.net/npm/d3@7"></script>
<script>
  const data = [5, 10, 3, 8, 2];

  const w = 500, h = 120;

  const svg = d3.select("#chart")
    .append("svg")
    .attr("width", w)
    .attr("height", h);

  svg.selectAll("circle")
    .data(data)
    .join("circle")
    .attr("cx", (d, i) => 50 + i * 90)
    .attr("cy", d => 100 - d * 8)
    .attr("r", d => d * 2);
</script>
```
